**Methane Emissions**
Global methane emissions are major part of global greenhouse gas emissions.The 20-year global warming potential of methane is 84.That is, over a 20-year period, it traps 84 times more heat per mass unit than carbon dioxide (CO2) and 32 times the effect when accounting for aerosol interactions. Global methane concentrations rose from 722 parts per billion (ppb) in pre-industrial times to 1866 ppb by 2019,an increase by a factor of 2.5 and the highest value in at least 800,000 years


[Mehtane emission data from the world bank website](https://data.worldbank.org/indicator/EN.ATM.METH.KT.CE?end=2012&start=1978&view=chart)

**Showing the emission of methane of every country (1970-2012)**

```python
def plot_countries(cleaned_dt):
    from pylab import rcParams
    rcParams['figure.figsize'] = (12,6)
    import numpy as np

    for i in range(0,len(cleaned_dt)):
        plt.plot(cleaned_dt.iloc[i,][2:-3])
    plt.xticks(range(len(range(1970,2012))),cleaned_dt.columns.values[2:-3], rotation = 70)
    
    plt.legend(cleaned_dt["Country Name"].values, ncol=4, loc='upper center', 
               bbox_to_anchor=[0.4, -0.1], 
               columnspacing=1.0, labelspacing=0.0,
               handletextpad=0.0, handlelength=1.0,
               fancybox=True)
    plt.ylabel("Methane Emission (kilo ton/capita)")
    plt.title("Methane Emission Per Capita by Country and Year")
    plt.show()
```

```python
plot_countries(cleaned_dt)  #plotting methane emission per capita for every country in the data
```

**Creating a function to select a specifc country and it's data**

```python
def filter_country(search_para):
    return cleaned_dt[cleaned_dt['Country Name'].str.contains(search_para,case=False)\
                     | cleaned_dt['Country Code'].str.contains(search_para,case=False)\
                     | cleaned_dt['Region'].str.contains(search_para,case=False)]
```

```python
filter_country('India')
```

```python
plot_countries(filter_country('South Asia')) #plotting methane emission of a specific region using the filter_function
```

**Creating an interactive UI widget**

```python
from IPython.display import display
# imports the widgets interact function
from IPython.html.widgets import interact

# decorate the search_and_plot_medals function using interact
# to create text field UI for search term
@interact(search='')
def interactive_plot_search_country(search):
    '''
    Plot and display data of country searched.
    '''
    filtered_country = filter_country(search)
    if len(filtered_country) == 0:
        print ('No hits')
        return
    
    # show plot 
    plot_countries(filtered_country)
    # output results
    display(filtered_country)
```
